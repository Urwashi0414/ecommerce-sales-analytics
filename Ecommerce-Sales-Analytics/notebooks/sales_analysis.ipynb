```python
snapshot_date = df["Order Date"].max()
snapshot_date
```

```python
rfm = df.groupby("Customer ID").agg({
    "Order Date": lambda x: (snapshot_date - x.max()).days,
    "Order ID": "count",
    "Sales": "sum"
})
```

```python
rfm.columns = ["Recency", "Frequency", "Monetary"]

rfm.head()
```

```python
rfm.head()
```

```python
rfm["R_score"] = pd.qcut(rfm["Recency"], 5, labels=[5,4,3,2,1])
rfm["F_score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels=[1,2,3,4,5])
rfm["M_score"] = pd.qcut(rfm["Monetary"], 5, labels=[1,2,3,4,5])

rfm.head()
```

```python
rfm["RFM_Score"] = rfm["R_score"].astype(str) + rfm["F_score"].astype(str) + rfm["M_score"].astype(str)

rfm.head()
```

```python
rfm.sort_values("RFM_Score", ascending=False).head(10)
```

```python
def segment_customer(score):
    if score >= "555":
        return "Champions"
    elif score >= "445":
        return "Loyal Customers"
    elif score >= "334":
        return "Potential Loyalist"
    else:
        return "At Risk"

rfm["Segment"] = rfm["RFM_Score"].apply(segment_customer)

rfm.head()
```

```python
rfm["Segment"].value_counts()
```

```python
df["OrderMonth"] = df["Order Date"].dt.to_period("M")
```

```python
df["CohortMonth"] = df.groupby("Customer ID")["OrderMonth"].transform("min")
```

```python
df["CohortIndex"] = (
    (df["OrderMonth"].dt.year - df["CohortMonth"].dt.year) * 12 +
    (df["OrderMonth"].dt.month - df["CohortMonth"].dt.month)
)
```

```python
cohort_data = df.groupby(["CohortMonth","CohortIndex"])["Customer ID"].nunique()

cohort_table = cohort_data.reset_index().pivot(
    index="CohortMonth",
    columns="CohortIndex",
    values="Customer ID"
)

cohort_table.head()
```

```python
cohort_size = cohort_table.iloc[:,0]

retention = cohort_table.divide(cohort_size, axis=0)

retention.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.heatmap(retention, annot=True, fmt=".0%", cmap="Blues")

plt.title("Customer Retention Cohort Analysis")
plt.show()
```

```python
basket = df.groupby(["Order ID", "Sub-Category"])["Quantity"].sum().unstack().fillna(0)
```

```python
basket = (basket > 0).astype(int)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_items = apriori(basket, min_support=0.02, use_colnames=True)

rules = association_rules(frequent_items, metric="lift", min_threshold=1)

rules.head()
```
